```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

xs = np.arange(10)
ys = np.ones_like(xs) * np.sin(xs)+ 10

yss = [2 * xs + i/2 for i in range(10)]
x = 5
a = np.interp(x, xs, ys)
plt.figure()
plt.plot(xs, ys)

def zero(guess, y=None):
    return np.interp(guess, xs, ys) - np.interp(guess, xs, y)

ress = []
for i in range(10):
    res = newton(zero, 5, args=((yss[i],)))
    ress.append(res)

y_res = np.interp(ress, xs, ys)
for i in range(10):
    line = plt.plot(xs, yss[i])
    plt.plot(ress[i], y_res[i], 'o', color=line[0].get_color())
plt.xlim([2, 5])
plt.ylim([5, 15])
```
